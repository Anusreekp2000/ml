```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import mean_squared_error
```

```python
df =pd.read_csv('Salary_Data.csv')
df.head()
```

```python
# Initialize parameters
m = 0  # slope
b = 0  # intercept

# Hyperparameters
alpha = 0.01  # learning rate
epochs = 1000  # number of iterations

# Gradient Descent
for epoch in range(epochs):
    predictions = m * df['YearsExperience'] + b
    errors = predictions - df['Salary']
    
    m -= alpha * (1/len(df)) * np.sum(2 * df['YearsExperience'] * errors)
    b -= alpha * (1/len(df)) * np.sum(2 * errors)
```

```python
pred = m * df['YearsExperience'] + b
pred
```

```python
# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(df['Salary'], predictions)
print("Mean Squared Error (MSE):", mse)
```

```python
# Plot the results
plt.scatter(df['YearsExperience'], df['Salary'], label='Original data')
plt.plot(df['YearsExperience'], predictions, label='Linear Regression', color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Print the final parameters
print(f"Final slope (m): {m}")
print(f"Final intercept (b): {b}")
```
